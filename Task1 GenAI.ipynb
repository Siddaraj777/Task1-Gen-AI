```python
# ==============================
# GenAI Prompt Engineering Assignment
# LangChain Prompt Templates - All Tasks
# ==============================

# Install LangChain
!pip install -q langchain

from langchain.prompts import PromptTemplate, ChatPromptTemplate

# =========================================
# Task 1: Replace Hardcoded Prompt
# =========================================
print("TASK 1 OUTPUT")
template1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template1.format(topic="Artificial Intelligence"))
print("\n")


# =========================================
# Task 2: Multi-Input Prompt System
# =========================================
print("TASK 2 OUTPUT")
multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_cases = [
    ("AI", "beginners", "friendly"),
    ("Python", "kids", "fun"),
    ("Deep Learning", "engineers", "technical")
]

for topic, audience, tone in test_cases:
    print(multi_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
print("\n")


# =========================================
# Task 3: Prompt Variations Engine
# =========================================
print("TASK 3 OUTPUT")
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
print("\n")


# =========================================
# Task 4: ChatPromptTemplate System
# =========================================
print("TASK 4 OUTPUT")

role_messages = {
    "teacher": "You are a helpful teacher who explains topics clearly.",
    "interviewer": "You are an interviewer who asks challenging questions.",
    "motivator": "You are a motivator who inspires learning."
}

role = "teacher"
topic = "Neural Networks"

chat_template = ChatPromptTemplate.from_messages([
    ("system", role_messages[role]),
    ("user", "Explain {topic}")
])

chat_prompt = chat_template.format_messages(topic=topic)

for msg in chat_prompt:
    print(f"{msg.type.upper()}: {msg.content}")
print("\n")


# =========================================
# Task 5: Input Validation Layer
# =========================================
print("TASK 5 OUTPUT")

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

audience, tone = validate_inputs("student", "cool")
print("Validated:", audience, tone)
print("\n")


# =========================================
# Task 6: Prompt Generator App
# =========================================
print("TASK 6 OUTPUT")

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions on {topic} for {audience} in a {tone} style",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=style_templates[style]
    )

    return template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

print(generate_prompt(
    "Neural Networks",
    "beginner",
    "fun",
    "storytelling"
))
print("\n")


# =========================================
# Task 7: Template Reusability Test
# =========================================
print("TASK 7 OUTPUT")

reusable_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in easy steps"
)

topics = [
    "AI",
    "Python",
    "Machine Learning",
    "Deep Learning",
    "NLP"
]

for item in topics:
    print(reusable_template.format(topic=item))
```

```python
# ==========================================
# ALL TASKS WITH PROMPT + ANSWER GENERATION
# ==========================================

!pip install -q langchain transformers torch

from langchain.prompts import PromptTemplate, ChatPromptTemplate
from transformers import pipeline

# Load text generation model
generator = pipeline(
    "text-generation",
    model="gpt2",
    max_new_tokens=60
)

# ==========================================
# Task 1
# ==========================================
print("TASK 1")
task1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners."
)

prompt = task1.format(topic="Artificial Intelligence")
print("PROMPT:", prompt)
print("ANSWER:", generator(prompt)[0]["generated_text"])
print("="*80)


# ==========================================
# Task 2
# ==========================================
print("TASK 2")
task2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

cases = [
    ("AI", "beginners", "friendly"),
    ("Python", "kids", "fun"),
    ("Deep Learning", "engineers", "technical")
]

for topic, audience, tone in cases:
    prompt = task2.format(topic=topic, audience=audience, tone=tone)
    print("PROMPT:", prompt)
    print("ANSWER:", generator(prompt)[0]["generated_text"])
    print("-"*60)


# ==========================================
# Task 3
# ==========================================
print("TASK 3")
templates = [
    PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step."
    ),
    PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 interview questions about {topic}."
    ),
    PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story."
    )
]

for temp in templates:
    prompt = temp.format(topic="Machine Learning")
    print("PROMPT:", prompt)
    print("ANSWER:", generator(prompt)[0]["generated_text"])
    print("-"*60)


# ==========================================
# Task 4
# ==========================================
print("TASK 4")
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful teacher."),
    ("user", "Explain {topic}")
])

messages = chat_template.format_messages(topic="Neural Networks")

for msg in messages:
    print(msg.type.upper() + ":", msg.content)
print("="*80)


# ==========================================
# Task 5
# ==========================================
print("TASK 5")

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

audience, tone = validate_inputs("kids", "friendly")
print("Validated:", audience, tone)
print("="*80)


# ==========================================
# Task 6
# ==========================================
print("TASK 6")

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style."
    )

    return template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
    )

prompt = generate_prompt(
    "Neural Networks",
    "beginner",
    "fun",
    "storytelling"
)

print("PROMPT:", prompt)
print("ANSWER:", generator(prompt)[0]["generated_text"])
print("="*80)


# ==========================================
# Task 7
# ==========================================
print("TASK 7")
reusable = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in easy steps."
)

topics = ["AI", "Python", "NLP", "Deep Learning", "Prompt Engineering"]

for topic in topics:
    prompt = reusable.format(topic=topic)
    print("PROMPT:", prompt)
    print("ANSWER:", generator(prompt)[0]["generated_text"])
    print("-"*60)
```
